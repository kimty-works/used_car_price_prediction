```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))  # 더 큰 차트 사이즈 설정
sns.barplot(x="Owner_Type", y="Price", data=df_car, ci=None)  # ci=None은 불확실성 막대를 제거
plt.title('Number of Cars by Owner Type')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 'First'와 'Second' Owner_Type에 따른 Year 데이터 추출
year_first = df_car[df_car['Owner_Type'] == '0']['Vehicle_Age']
year_second = df_car[df_car['Owner_Type'] == '1']['Vehicle_Age']

# 독립표본 t-검정 수행
t_stat, p_value = ttest_ind(year_first, year_second)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
import matplotlib.pyplot as plt

# 데이터 프레임 정렬
df_car_sorted = df_car.sort_values('Vehicle_Age')

# 트렌드 차트 그리기, 선의 색 변경
plt.figure(figsize=(10, 5))  # 그래프 크기 설정
plt.plot(df_car_sorted['Vehicle_Age'], df_car_sorted['Price'], marker='o', linestyle='none', color='blue')  # 선 색상을 빨간색으로 설정
plt.title('Price Trend Over Vehicle_Age')  # 그래프 제목 설정
plt.xlabel('Vehicle_Age')  # x축 라벨 설정
plt.ylabel('Price')  # y축 라벨 설정
plt.grid(True)  # 그리드 추가
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# df_car 데이터프레임과 'Price', 'Year' 열이 이미 정의되어 있다고 가정

# Price에 대한 로그 변환 수행 및 log_price 열로 저장
df_car['log_price'] = np.log1p(df_car['Price'])

plt.figure(figsize=(10, 5))  # 그래프 크기 설정
# Year별 log_price의 boxplot 그리기
df_car.boxplot(column='log_price', by='Vehicle_Age', figsize=(10,5))
plt.title('Log-Price Distribution by Year')  # 그래프 제목 설정
plt.suptitle('')  # 기본 제목 제거
plt.xlabel('Vehicle_Age')  # x축 라벨 설정
plt.ylabel('Log-Price')  # y축 라벨 설정 (변경된 부분)
plt.xticks(rotation=45)  # x축 라벨 회전
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 프레임 정렬
df_car_sorted = df_car.sort_values('Owner_Type')

# 트렌드 차트 그리기, 선의 색 변경
plt.figure(figsize=(10, 5))  # 그래프 크기 설정
plt.plot(df_car_sorted['Owner_Type'], df_car_sorted['Price'], marker='o', linestyle='none', color='blue')  # 선 색상을 빨간색으로 설정
plt.title('Price Trend on Owner_Type')  # 그래프 제목 설정
plt.xlabel('Owner_Type')  # x축 라벨 설정
plt.ylabel('Price')  # y축 라벨 설정
plt.grid(True)  # 그리드 추가
plt.show()
```

```python
# 소유주별 연식별 가격 히트맵
df_agg = df_car.groupby(["Owner_Type", "Vehicle_Age"], as_index = False).agg("mean")
df_pivot = df_agg.pivot("Owner_Type", "Vehicle_Age", "Price")
sns.heatmap(df_pivot, cmap="Blues")
```

```python
import seaborn as sns

# 산점도와 선형 회귀선 그리기
plt.figure(figsize=(8, 6))
sns.regplot(x=df_car['Kilometers_Driven'], y=df_car['Vehicle_Age'], color='blue')
plt.title('Scatter Plot with Linear Regression Line of Kilometers Driven vs. Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Vehicle_Age')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# log_price와 owner_type의 빈도 분석
cross_tab = pd.crosstab(df_car['log_price'], df_car['Owner_Type'])

# 카이제곱 검정 수행
chi2, p_val, _, _ = chi2_contingency(cross_tab)

print("Chi-square statistic:", chi2)
print("P-value:", p_val)
```

```python
##Year 변수의 정규성 테스트

from scipy.stats import normaltest

stat, p = normaltest(df_car['Vehicle_Age'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

```python
from scipy.stats import spearmanr, kendalltau

# 스피어만의 순위 상관계수 계산
spearman_corr, spearman_p = spearmanr(df_car['Vehicle_Age'], df_car['Price'])
print(f"Spearman's correlation coefficient: {spearman_corr:.3f}, P-value: {spearman_p:.3f}")

# 켄달의 타우 상관계수 계산
kendall_corr, kendall_p = kendalltau(df_car['Vehicle_Age'], df_car['Price'])
print(f"Kendall's tau correlation coefficient: {kendall_corr:.3f}, P-value: {kendall_p:.3f}")
```
